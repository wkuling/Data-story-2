```python
grades = grades[grades["ONDERWIJSTYPE VO"]=="VWO"]
```

```python
temp=grades[[u'INSTELLINGSNAAM VESTIGING_x', u'VAKNAAM', "GEM. CIJFER CIJFERLIJST_x", u'AANTAL CENTRALE EXAMENS MEETELLEND VOOR DIPLOMA_x'
             , "GEM. CIJFER CIJFERLIJST_y", u'AANTAL CENTRALE EXAMENS MEETELLEND VOOR DIPLOMA_y']]

gymnasium = grades["INSTELLINGSNAAM VESTIGING_x"].str.contains("Gymn")

temp=temp[gymnasium]
temp=temp.groupby('VAKNAAM').mean()
temp["AANTAL CENTRALE EXAMENS MEETELLEND VOOR DIPLOMA_x"] = grades[[u'VAKNAAM', "AANTAL CENTRALE EXAMENS MEETELLEND VOOR DIPLOMA_x"]][gymnasium].groupby('VAKNAAM').sum()
temp["AANTAL CENTRALE EXAMENS MEETELLEND VOOR DIPLOMA_y"] = grades[[u'VAKNAAM', "AANTAL CENTRALE EXAMENS MEETELLEND VOOR DIPLOMA_y"]][gymnasium].groupby('VAKNAAM').sum()
temp = temp[temp["AANTAL CENTRALE EXAMENS MEETELLEND VOOR DIPLOMA_x"]!=0]
temp.columns = ["Average Grade 2013", "Number of exams 2013", "Average Grade 2012", "Number of exams 2012"]
tempgym = temp
tempgym
```

```python
temp=grades[[u'INSTELLINGSNAAM VESTIGING_x', u'VAKNAAM', "GEM. CIJFER CIJFERLIJST_x", u'AANTAL CENTRALE EXAMENS MEETELLEND VOOR DIPLOMA_x'
             , "GEM. CIJFER CIJFERLIJST_y", u'AANTAL CENTRALE EXAMENS MEETELLEND VOOR DIPLOMA_y']]
temp=temp[~gymnasium]
temp=temp.groupby('VAKNAAM').mean()
temp["AANTAL CENTRALE EXAMENS MEETELLEND VOOR DIPLOMA_x"] = grades[[u'VAKNAAM', "AANTAL CENTRALE EXAMENS MEETELLEND VOOR DIPLOMA_x"]][~gymnasium].groupby('VAKNAAM').sum()
temp["AANTAL CENTRALE EXAMENS MEETELLEND VOOR DIPLOMA_y"] = grades[[u'VAKNAAM', "AANTAL CENTRALE EXAMENS MEETELLEND VOOR DIPLOMA_y"]][~gymnasium].groupby('VAKNAAM').sum()
temp = temp[temp["AANTAL CENTRALE EXAMENS MEETELLEND VOOR DIPLOMA_x"]!=0]
temp.columns = ["Average Grade 2013", "Number of exams 2013", "Average Grade 2012", "Number of exams 2012"]
tempnotgym = temp
tempnotgym
```

```python
compare = pd.concat([tempgym, tempnotgym], axis=1, join='inner')
compare.columns = ["Grade_2013_gym", "#exams_2013_gym", "Grade_2012_gym", "#exams_2012_gym",
                   "Grade_2013_notgym", "#exams_2013_notgym", "Grade_2012_notgym", "#exams_2012_notgym"]
compare["Diff_2013"]=compare["Grade_2013_gym"]-compare["Grade_2013_notgym"]
compare["Diff_2012"]=compare["Grade_2012_gym"]-compare["Grade_2012_notgym"]
compare[["Diff_2013", "#exams_2013_gym", "#exams_2013_notgym", "Diff_2012", "#exams_2012_gym", "#exams_2012_notgym"]]
```
